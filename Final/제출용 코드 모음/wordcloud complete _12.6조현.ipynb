```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
from wordcloud import STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
import matplotlib.pyplot as plt
```

## 1. 전처리와 감성분석 완료된 csv파일 불러오기 

```python
bit_senti = pd.read_csv(r"C:\Users\Lenovo\OneDrive\빅데이터분석&활용팀플\csv outputs\tweets_sentiment_ujw.csv")
bit_senti.tail()
```

### 각 트윗 토큰화

```python
# 토큰화 시도 (나름대로 성공했으나 이 코드는 쓰지 않음)
cleantext = list(bit_senti['cleaned_tweets'])

for i in cleantext:
    result  =[]
    word_tokens = word_tokenize(str(i))
    print('\n', '토큰화: ',word_tokens)  #각 트윗 토크나이즈까지 완료
    wordcloud = WordCloud().generate(i)
```

## 2. 워드클라우드 실행코드 - 대략적인 분기별로 나누어 확인하기

```python
#1 30만~40만 행씩 나누어서 워드클라우드 진행
comment_words = ''
stopwords = set(STOPWORDS)

for val in bit_senti['cleaned_tweets'][:300000]:
    # typecaste each val to string
    val = str(val)
 
    # split the value
    tokens = val.split()
     
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
     
    comment_words += " ".join(tokens)+" "

wc = WordCloud(width = 800, height = 800, max_words=1000,background_color ='white', stopwords = stopwords, min_font_size = 10).generate(comment_words)

 
# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wc)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
