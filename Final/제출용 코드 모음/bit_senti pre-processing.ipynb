## 정규표현식으로 날짜가 아닌 값은 삭제

```python
import re  #정규표현식
```

```python
from tqdm import tqdm
import progressbar

for i in tqdm(range(80, 100)):
    dt_regex = re.compile('[a-z]', re.I)
    valid = dt_regex.search(aa['date'][i])
    if valid:
        aa.drop(index =i ,axis =0, inplace = True)
    else:
        pass

print(aa)
```
